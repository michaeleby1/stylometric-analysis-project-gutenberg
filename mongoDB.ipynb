```python
for doc in collection.find({'name': 'John Doe'}):
    doc['address'] = nlp('32421 elm street')
    print(doc)
```

```python
for doc in collection.find({'name': 'John Doe'}):
    doc['address'] = nlp('321 elm street')
#     print(doc['address'])
    for ent in doc['address'].ents:
        print(ent.text, ent.label_)
```

```python
for doc in collection.find({'age': {'$gt': 20}}):
    print(doc)
```

```python
collection.update_one({'name': 'John Doe'}, {'$set': {'age': 29}})
for doc in collection.find({'name': 'John Doe'}):
    print(doc)
```
